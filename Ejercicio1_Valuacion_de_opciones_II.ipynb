#### Ejercicio 

Estudiar el comportamiento de las trayectorias al variar $\mu$ y $\sigma$.

```python
from random import gauss
from math import exp
import matplotlib.pyplot as plt
```

```python
mu=0.001 
sigma=0.04  #sigma
p0=10
t=200
```

```python
def trayectoria(mu,sigma,p0,t):
    l=[p0]
    pt=p0
    for i in range(t):
        pt=pt*exp(mu + sigma*gauss(0,1))
        l.append(pt)
    return l
```

```python
for i in range(10):
    l=trayectoria(mu,sigma,p0,t)
    plt.plot(l)
plt.show()
```

### Con $\mu$ ms pequea

```python
mu=0.0001 
sigma=0.04  #sigma
p0=10
t=200
for i in range(10):
    l=trayectoria(mu,sigma,p0,t)
    plt.plot(l)
plt.show()
```

### Con $\mu$ y $\sigma$ ms pequeas

```python
mu=0.0001 
sigma=0.004  #sigma
p0=10
t=200
for i in range(10):
    l=trayectoria(mu,sigma,p0,t)
    plt.plot(l)
plt.show()
```

Se observa que entre ms pequeos los valores, las cotas se cierran ms.

### Con $\mu$ y $\sigma$ ms grandes

```python
mu=0.01 
sigma=0.4  #sigma
p0=10
t=200
for i in range(10):
    l=trayectoria(mu,sigma,p0,t)
    plt.plot(l)
plt.show()
```

Los precios se disparan, recordando que se esperan de orden $10^{-3}$
