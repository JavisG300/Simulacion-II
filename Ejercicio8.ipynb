$1. \int_{0}^{1} (1-x^{2})^{\frac{3}{2}}dx$

```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
def integral_mc(n):
    I = 0
    lx = []
    ly = []
    counter = 0
    for i in range(n):
        x = random.random()
        y = random.random()
        if y < (1 - x**2)**(3/2):
            counter = counter + 1
        i=i+1
        I = counter/i
        lx.append(i)
        ly.append(I)
    print("Valor de la integral =", I)
    
    fig = plt.figure(1, figsize=(8, 8))        
    plt.xlabel("Iteraciones")
    plt.ylabel("EstimaciÃ³n de la integral")
    plt.grid()
    plt.plot(lx, ly, color="blue", markersize=1)
    plt.show()
```

```python
print(integral_mc(1000))
```

2. $\int_{-2}^{2}exp(x+x^{2})dx$

```python
def integral_mc(n):
    I = 0
    lx = []
    ly = []
    counter = 0
    for i in range(n):
        x = random.uniform(-2,2)
        y = random.uniform(-2,2)
        if y < np.exp(x + x**2):
            counter = counter + 1
        i=i+1
        I = counter/i
        lx.append(i)
        ly.append(I)
    print("Valor de la integral =", I)
    
    fig = plt.figure(1, figsize=(8, 8))        
    plt.xlabel("Iteraciones")
    plt.ylabel("EstimaciÃ³n de la integral")
    plt.grid()
    plt.plot(lx, ly, color="green", markersize=1)
    plt.show()
```

```python
print(integral_mc(1000))
```
