## Ejercicio 6
Repetir el experimento usando un cuarto de elipse

```python
N = 1000
def pi_Taylor(n):
  l = []
  m = 0
  for i in range(n):
    x = random.random()
    y = random.random()
    z = ((x**2)+(y**2))**(1/2)
    if z < 1:
        m = m+1
    pi = 4*(m/(i+1))
    print(i,pi)
    l.append(pi) 
  plt.plot(l, color="blue", markersize=1)
  plt.xlabel("Iteraciones")
  plt.ylabel("Estimaci√≥n de $\pi$")
  plt.grid()
  plt.show()
pi_Taylor(N)
```
